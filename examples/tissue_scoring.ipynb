# Example: Tissue scoring

#### Load gene annotation:

If you do not have a gene annotation yet, you can download automatically from BioMart via

```python
from PascalX.genome import genome

G = genome()
G.get_ensembl_annotation('biomart_GRCh38.tsv')
```

The annotation needs to be loaded into a tissue scorer as follows:

```python
from PascalX.genexpr import genexpr

GE = genexpr()
GE.load_genome('biomart_GRCh38.tsv')
```

#### Import GTEx data:

The tissue scorer requires GTEx data. It can be downloaded automatically via the following command:

```python
GE.get_GTEX_expr('yourGTEXfilename')
```

Note that you only need to download and import once. The data will be stored in a ``GTEX/`` subfolder. The import will take several hours.

#### Load GTEx data:

```python
GE.load_expr('yourGTEXfilename')
```

#### Visualization:

```python
GE.plot_genexpr(['AOAH','CSMD1','CDH13'],tzscore=True,cbar_pos=(0.0, 0.1, 0.01, 0.5))
```

#### Scoring:

```python
R = GE.chi2rank([ ['Pathway 1',['AOAH','CSMD1','CDH13'] ] ])
```
